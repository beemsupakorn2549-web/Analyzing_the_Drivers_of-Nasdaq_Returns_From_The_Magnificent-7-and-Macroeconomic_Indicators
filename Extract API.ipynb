```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

### Select Only Magnificent 7 Stock In Nasdaq

```python
tickers = [
    'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'TSLA', # Magnificent 7 stocks
    '^IXIC',      # Nasdaq Composite Index
    'DX-Y.NYB',   # Dollar Index
    '^VIX',       # VIX Index
    'CL=F',       # Crude Oil
    'GC=F',       # Gold
    'TIP'         # Treasury Inflation-Protected Securities
]
```

### Using Last 5 Years Data

```python
start_date = "2021-01-01"
end_date = "2025-12-31"
print(f"Extracting data from {start_date} to {end_date} ...")
```

### Extract Data From Nasdaq by yfinance (Yahoo Finance)

```python
try:
    data = yf.download(tickers, start=start_date, end=end_date)
    
    if data.empty:
        print("\n!!! No data found: Please check the date range or internet connection !!!")
    else:
        # Choose only 'Close' prices immediately (which are adjusted prices when auto_adjust=True)
        # yfinance returns a DataFrame with stock names as columns
        df = data['Close']

        print("\n--- head 5 ---")
        print(df.head())
        
        print("\n--- Check data shape ---")
        print(f"Row: {len(df)}")
        print(f"Column: {len(df.columns)}")

except Exception as e:
    print(f"\n Invaild Extract Data: {e}")
```

### Change column names to be MLR-friendly

```python
rename_dict = {
    '^IXIC': 'Nasdaq',
    '^TNX': 'US_10Y_Bond',
    'DX-Y.NYB': 'Dollar_Index',
    '^VIX' : 'VIX_Fear_Index',
    'CL=F': 'WTI_Crude_Oil',
    'GC=F': 'COMEX_Gold',
    'TIP': 'Treasury_Inflation'
}

df = df.rename(columns=rename_dict)
df
```

### Save Data to csv

```python
df.to_csv('mag7_nasdaq.csv')
print("\n Complete Save: 'mag7_nasdaq.csv'")
```
