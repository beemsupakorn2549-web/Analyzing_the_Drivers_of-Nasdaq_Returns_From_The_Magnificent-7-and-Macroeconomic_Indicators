### Feature Engineering
#### We will analyze the "percentage change in price" (Returns), not the "raw price" Because the raw price has a trend (non-stationary), making the statistics easily distorted.

```python
df_returns = df.drop(columns=['Date_Cleaned']).pct_change().dropna()
df_returns['Date_Cleaned'] = df['Date_Cleaned'].iloc[1:].values
```

### Save percentage change in price Flie
